```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already loaded your data
data = pd.read_csv('fatalities.csv')

# Convert date columns to datetime format
data['date_of_event'] = pd.to_datetime(data['date_of_event'])
data['date_of_death'] = pd.to_datetime(data['date_of_death'])

# Group by date and count the occurrences
events = data.groupby('date_of_event').size().reset_index(name='Number of Events')
deaths = data.groupby('date_of_death').size().reset_index(name='Number of Deaths')

# Combine the data for Events and Deaths
combined_data = pd.merge(events, deaths, how='outer', left_on='date_of_event', right_on='date_of_death')

# Use Seaborn to create a scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x='date_of_event', y='Number of Events', data=combined_data, label='Number of Events', marker='o')
sns.scatterplot(x='date_of_death', y='Number of Deaths', data=combined_data, label='Number of Deaths', marker='D')

plt.title('Fatalities Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame
np.random.seed(42)
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D', 'E'], size=100),
    'Value': np.random.randint(1, 100, size=100),
    'Date': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'Group': np.random.choice(['X', 'Y'], size=100),
    'Binary': np.random.choice([0, 1], size=100),
}

df = pd.DataFrame(data)
print(df)
# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Value', y='Binary', data=df, hue='Category', style='Group')
plt.title('Scatter Plot')
plt.show()

# Line Plot
plt.figure(figsize=(8, 5))
sns.lineplot(x='Date', y='Value', data=df)
plt.title('Line Plot')
plt.show()

# Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=df, ci=None)
plt.title('Bar Plot')
plt.show()

# Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Category', y='Value', data=df, hue='Group')
plt.title('Box Plot')
plt.show()

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data['Value'], bins=20, kde=True)
plt.title('Histogram')
plt.show()

# Pie Chart (using Matplotlib)
plt.figure(figsize=(8, 5))
plt.pie(df.groupby('Category')['Value'].sum(), labels=df['Category'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()

# Stacked Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=df, hue='Group', ci=None)
plt.title('Stacked Bar Chart')
plt.show()
```
